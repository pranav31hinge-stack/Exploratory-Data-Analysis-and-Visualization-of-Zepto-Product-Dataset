```python

"""
Zepto Dataset - Complete Data Analysis and Visualization (No Path Errors)
-----------------------------------------------------------------------
Performs:
1. Load + Inspect
2. Clean Missing Data
3. Info & Summary
4. Visualization Techniques
5. Auto-save cleaned dataset safely
"""

# === Import Libraries ===
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_theme(style="whitegrid", palette="muted")

# === Step 1: Set up file paths ===
# Modify this to your actual dataset path if needed
input_file = "zepto_v1.xlsx"   # Your Excel file should be in the same folder as this script
output_dir = os.path.join(os.getcwd(), "data_output")  # Save cleaned data here
os.makedirs(output_dir, exist_ok=True)  # Create folder if not exists

# === Step 2: Load Dataset ===
df = pd.read_excel(input_file)
print("=== DATASET INFORMATION ===")
print(f"Shape: {df.shape}")
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())

# === Step 3: Clean Missing Data ===
df_cleaned = df.dropna().copy()
print("\nâœ… Missing data cleaned successfully.")
print(f"New shape: {df_cleaned.shape}")

# === Step 4: Statistical Summary ===
print("\n=== STATISTICAL SUMMARY ===")
print(df_cleaned.describe(include='all'))

# === Step 5: Correlation Heatmap ===
plt.figure(figsize=(10, 6))
corr = df_cleaned.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Columns", fontsize=14)
plt.tight_layout()
plt.show()

# === Step 6: Pairplot ===
numeric_cols = df_cleaned.select_dtypes(include=['int64', 'float64'])
if numeric_cols.shape[1] > 1:
    sns.pairplot(numeric_cols)
    plt.suptitle("Pairplot of Numeric Columns", y=1.02)
    plt.show()

# === Step 7: Data Visualization Techniques ===

## 1ï¸âƒ£ Histogram: Distribution of numerical features
for col in numeric_cols.columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df_cleaned[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

## 2ï¸âƒ£ Boxplots: Detecting outliers
for col in numeric_cols.columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df_cleaned[col])
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()

## 3ï¸âƒ£ Bar Plot: Average discounted selling price by category
if 'Category' in df_cleaned.columns and 'discountedSellingPrice' in df_cleaned.columns:
    plt.figure(figsize=(10, 6))
    category_mean = df_cleaned.groupby('Category')['discountedSellingPrice'].mean().sort_values(ascending=False)
    sns.barplot(x=category_mean.values, y=category_mean.index)
    plt.title("Average Discounted Selling Price by Category")
    plt.xlabel("Average Discounted Selling Price")
    plt.ylabel("Category")
    plt.show()

## 4ï¸âƒ£ Countplot: Number of products per category
if 'Category' in df_cleaned.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(y='Category', data=df_cleaned, order=df_cleaned['Category'].value_counts().index)
    plt.title("Number of Products per Category")
    plt.xlabel("Count")
    plt.ylabel("Category")
    plt.show()

## 5ï¸âƒ£ Scatter Plot: MRP vs Discounted Price
if {'mrp', 'discountedSellingPrice'}.issubset(df_cleaned.columns):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='mrp', y='discountedSellingPrice', data=df_cleaned, hue='Category', alpha=0.7)
    plt.title("MRP vs Discounted Selling Price by Category")
    plt.xlabel("MRP")
    plt.ylabel("Discounted Selling Price")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

## 6ï¸âƒ£ Jointplot: Relationship between weight and price
if {'weightInGms', 'discountedSellingPrice'}.issubset(df_cleaned.columns):
    sns.jointplot(x='weightInGms', y='discountedSellingPrice', data=df_cleaned, kind='hex', color='green')
    plt.suptitle("Weight (gms) vs Discounted Selling Price", y=1.02)
    plt.show()

## 7ï¸âƒ£ Heatmap for Category vs Out-of-Stock Count
if {'Category', 'outOfStock'}.issubset(df_cleaned.columns):
    pivot = pd.crosstab(df_cleaned['Category'], df_cleaned['outOfStock'])
    plt.figure(figsize=(8, 5))
    sns.heatmap(pivot, annot=True, cmap="YlGnBu", fmt='d')
    plt.title("Out of Stock Count by Category")
    plt.show()

# === Step 8: Save Cleaned Data ===
output_csv = os.path.join(output_dir, "zepto_v1_cleaned.csv")
output_excel = os.path.join(output_dir, "zepto_v1_cleaned.xlsx")

df_cleaned.to_csv(output_csv, index=False)
df_cleaned.to_excel(output_excel, index=False)

print("\nðŸ’¾ Cleaned dataset saved successfully:")
print(f" - CSV: {output_csv}")
print(f" - Excel: {output_excel}")

# === Step 9: Display First Few Rows ===
print("\n=== PREVIEW OF CLEANED DATASET ===")
print(df_cleaned.head())
```
